```python
def set_limit(bank, mfo, loan, limit, percentage):
    if loan > 50:
        return 0
    elif (bank > 15000) or (mfo > 7000):
        return 5000
    elif (loan > 15) and (mfo > 5000):
        return limit
    elif (bank < 1000) and (mfo < 8000) and (loan < 20):
        return limit + 10*percentage
    else:
        return 2000
```

```python
df['new_limit'] = df.apply(lambda x: set_limit(x.Просрочка_в_банке, x.Просрочка_в_МФО, x.Просрочка_в_нашем_прошлом_займе, x.Лимит_из_последнего_закрытого_займа, x.Оплаченные_проценты_в_прошлом_займе), axis=1)
```
