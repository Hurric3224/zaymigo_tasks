```python
# функция для преобразования формата даты


def split_date(date_info):
    return date_info.split('-')[0] + ' ' + date_info.split('-')[1]
df['date'] = df['date'].apply(split_date)
date_map = {'2020 Aug': '2020/08',
            '2019 Sep': '2019/09',
            '2019 Oct': '2019/10',
            '2019 Aug': '2019/08',
            '2019 Nov': '2019/11',
            '2019 Jul': '2019/07',
            '2019 Apr': '2019/04',
            '2019 Dec': '2019/12',
            '2020 May': '2020/05',
            '2020 Jun': '2020/06',
            '2019 May': '2019/05',
            '2020 Mar': '2020/03',
            '2020 Jan': '2020/01',
            '2019 Jun': '2019/06',
            '2020 Jul': '2020/07',
            '2020 Apr': '2020/04',
            '2020 Feb': '2020/02',
            '2019 Mar': '2019/03'
            }
df['date'] = df['date'].map(date_map)
```

```python
df2 = df.groupby(['date', 'is_client_repeated', 'NPL1'], as_index=False).agg({
        'request_id': 'count',
        'loan_period': 'mean',
        'loan_amount': 'sum',
        'paid_total': 'sum',
        'limit': 'mean',       
        }).round()
```

```python
df2.rename(columns={'request_id': 'requests_count'}, inplace=True)
df2.head()
```

```python
df3 = df.groupby(['date', 'is_client_repeated'], as_index=False).agg({
        'request_id': 'count',
        'loan_amount': 'sum'
        }).round()
df3.head()
```

```python
df2 = df2.merge(df3, on=['date', 'is_client_repeated'], how='left')
df2.head()
```

```python
df2['NPL1, %'] = (df2['requests_count']/df2['request_id']*100).round(1)
df2['ROI, %'] = ((df2['paid_total']-df2['loan_amount_x'])/df2['loan_amount_x']*100).round(1)
df2['NPL1, % от ОД'] = ((df2['loan_amount_x']/df2['loan_amount_y'])*100).round(1)
```

```python
df2.head()
```

```python
report = df2
report = report[['date', 'requests_count', 'NPL1, %', 'NPL1, % от ОД', 'ROI, %', 'is_client_repeated', 'NPL1']]
report.rename(columns={'date': 'Дата',
                         'is_client_repeated': 'Тип клиента',
                         'requests_count': 'Количество'
                        }, inplace=True)
client_map = {0: 'Первчика',
              1: 'Повторники'}
report['Тип клиента'] = report['Тип клиента'].map(client_map)
report = report[['Дата', 'Тип клиента', 'Количество', 'NPL1', 'NPL1, %', 'NPL1, % от ОД', 'ROI, %']]
report = report.sort_values(by=['Дата', 'Тип клиента'])
report.head()
```
